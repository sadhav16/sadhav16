  Implementing ***Linear Regression*** model


**1.Importing the required libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
```

**2.Load the data**


```python
data=pd.read_csv('Advertising.csv')
```

```python
data.head()
```

```python
data.corr()['Sales']
```

**3.Setting the training features and predicting feature**


```python
X=data[['TV','Radio','Newspaper']].to_numpy()
X.shape
```

```python
Y=data['Sales'].to_numpy()
Y.shape
```

```python
X
```

**4.Importing the required model and training the model with the data**


```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X,Y)
pred=model.predict(X)
pred.shape
```

```python
pred
```

**5.Calculating the error percentage**

```python
from sklearn.metrics import mean_absolute_percentage_error,accuracy_score,mean_squared_error
mean_absolute_percentage_error(Y,pred)
```

```python
mean_squared_error(Y,pred)
```

**6.Model is trained and ready to predict the required values**

```python
model.predict([[230.1,  37.8,  69.2]]), Y[0]
```
